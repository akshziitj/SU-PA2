```python
import os
import torch
import torchaudio
import torchaudio.transforms as T
from torch.utils.data import Dataset, DataLoader
from pathlib import Path
from pydub import AudioSegment
import librosa
import numpy as np
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm
from transformers import Wav2Vec2FeatureExtractor, UniSpeechSatForXVector
from peft import LoraConfig, get_peft_model
from sklearn.metrics import roc_curve, auc

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained('microsoft/unispeech-sat-base-plus-sv')
base_model = UniSpeechSatForXVector.from_pretrained('microsoft/unispeech-sat-base-plus-sv').to(device)
```

#### Dataset creation

```python
class VoxCelebDataset(Dataset):
    def __init__(self, data_folder, speaker_list, max_audios=20):
        self.data = []
        self.speaker_to_idx = {speaker: idx for idx, speaker in enumerate(speaker_list)}

        for speaker in speaker_list:
            speaker_folder = Path(data_folder) / speaker
            audio_files = list(speaker_folder.rglob("*.m4a"))

            if len(audio_files) > max_audios:
                audio_files = random.sample(audio_files, max_audios)

            for audio_file in audio_files:
                wav_path = audio_file.with_suffix('.wav')
                if not wav_path.exists():
                    audio = AudioSegment.from_file(audio_file)
                    audio.export(wav_path, format="wav")

                self.data.append((self.speaker_to_idx[speaker], wav_path))  # Convert speaker to index

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        speaker_idx, audio_path = self.data[idx]
        embedding = self.extract_embedding(audio_path)
        return torch.tensor(embedding, dtype=torch.float32), torch.tensor(speaker_idx, dtype=torch.long)  # Convert to tensor

    def extract_embedding(self, audio_path):
        waveform, sr = librosa.load(str(audio_path), sr=16000)
        inputs = feature_extractor(waveform, sampling_rate=16000, return_tensors="pt")
        inputs = {key: val.to(device) for key, val in inputs.items()}
        with torch.no_grad():
            embeddings = model(**inputs).embeddings
        return embeddings.cpu().numpy().flatten()
```

#### ArcFace Loss

```python
import math

class ArcFaceLoss(nn.Module):
    """
    ArcFace Loss as described in the paper
    "ArcFace: Additive Angular Margin Loss for Deep Face Recognition"
    """
    def __init__(self, in_features, out_features, s=30.0, m=0.4, easy_margin=False):
        super(ArcFaceLoss, self).__init__()
        self.in_features = in_features
        self.out_features = out_features
        self.s = s  # scale factor
        self.m = m  # margin

        # Weight initialization
        self.weight = nn.Parameter(torch.FloatTensor(out_features, in_features))
        nn.init.xavier_uniform_(self.weight)

        self.easy_margin = easy_margin
        self.cos_m = math.cos(m)
        self.sin_m = math.sin(m)
        self.th = math.cos(math.pi - m)  # Threshold for margin application
        self.mm = math.sin(math.pi - m) * m  # Adjustment for angles beyond the threshold

    def forward(self, input, label):
        # Normalize input and weight
        cosine = F.linear(F.normalize(input), F.normalize(self.weight))
        cosine = torch.clamp(cosine, -1 + 1e-7, 1 - 1e-7)  # stability

        # Compute sine(theta) and phi(theta)
        sine = torch.sqrt(1.0 - torch.pow(cosine, 2))
        phi = cosine * self.cos_m - sine * self.sin_m  # cos(tehta+m) formula

        # Apply easy margin if needed
        if self.easy_margin:
            phi = torch.where(cosine > 0, phi, cosine)
        else:
            phi = torch.where(cosine > self.th, phi, cosine - self.mm)

        # Convert labels to one-hot encoding
        one_hot = F.one_hot(label, num_classes=self.out_features).float()

        # Compute final logits
        output = (one_hot * phi) + ((1.0 - one_hot) * cosine)
        output *= self.s  # Apply scale factor

        return output  # These logits will be passed to nn.CrossEntropyLoss()

```

#### Training and Validation Loop

```python
from tqdm import tqdm


def train(model, arcface, train_dataloader, val_dataloader, optimizer, criterion, device, epochs=10):
    for epoch in range(epochs):
        model.train()
        total_train_loss = 0
        correct_train = 0
        total_train = 0

        # Training phase
        for batch_idx, (embeddings, labels) in enumerate(tqdm(train_dataloader, desc=f"Epoch {epoch+1} [Training]")):
            embeddings, labels = embeddings.to(device), labels.to(device)

            # Reshape embeddings - UniSpeech typically outputs [batch, seq_len, hidden_size]
            if embeddings.dim() > 2:
                embeddings = embeddings.mean(dim=1)  # Mean pooling over sequence length

            optimizer.zero_grad()

            # Forward pass through ArcFace
            logits = arcface(embeddings, labels)
            loss = criterion(logits, labels)

            # Compute training accuracy
            _, predicted = torch.max(logits.data, 1)
            total_train += labels.size(0)
            correct_train += (predicted == labels).sum().item()
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            optimizer.step()

            total_train_loss += loss.item()

        train_loss = total_train_loss / len(train_dataloader)


        # Validation phase
        model.eval()
        val_loss = 0
        correct = 0
        total = 0

        with torch.no_grad():
            for val_embeddings, val_labels in tqdm(val_dataloader, desc=f"Epoch {epoch+1} [Validation]"):
                val_embeddings, val_labels = val_embeddings.to(device), val_labels.to(device)

                if val_embeddings.dim() > 2:
                    val_embeddings = val_embeddings.mean(dim=1)

                val_logits = arcface(val_embeddings, val_labels)
                batch_loss = criterion(val_logits, val_labels)
                val_loss += batch_loss.item()

                # Compute validation accuracy
                _, predicted = torch.max(val_logits.data, 1)
                total += val_labels.size(0)

        val_loss = val_loss / len(val_dataloader)


        print(f"\nEpoch {epoch+1} Results:")
        print(f"Training Loss: {train_loss:.4f}")
        print(f"Validation Loss: {val_loss:.4f}")
```

```python
import soundfile as sf
from pydub import AudioSegment
from pathlib import Path

def mix_audio(file1, file2, output_path, snr=0):
    """
    Mix two audio files with a given Signal-to-Noise Ratio (SNR).
    """
    y1, sr1= librosa.load(file1, sr=16000)
    y2, sr2= librosa.load(file2, sr=16000)

    assert sr1== sr2, "Sampling rates must be the same"

    len_min= min(len(y1), len(y2))
    y1, y2= y1[:len_min], y2[:len_min]

    # Normalize
    y1 = y1 / np.max(np.abs(y1))
    y2 = y2 / np.max(np.abs(y2))

    # Scale second speaker using SNR
    scale = 10 ** (-snr / 20)
    mixed_audio = y1 + scale * y2
    mixed_audio = mixed_audio / np.max(np.abs(mixed_audio))  # Normalize

    sf.write(output_path, mixed_audio, sr1)
    return output_path
```

#### Resample for Sepformer

```python
# Function to resample audio to target sample rate
def resample_audio(audio_path, target_sr=8000):
   #sepformer needs 8000 sr
    waveform, sr = torchaudio.load(audio_path)
    if sr != target_sr:
        resampler = torchaudio.transforms.Resample(sr, target_sr)
        waveform = resampler(waveform)
    return waveform.squeeze(), target_sr
```

#### Evaluate Separation

```python
# Function to evaluate separation performance
def evaluate_separation(reference_sources, estimated_sources, fs=8000):
    """
    Compute metrics for source separation evaluation:
    - SDR (Signal to Distortion Ratio)
    - SIR (Signal to Interference Ratio)
    - SAR (Signal to Artifacts Ratio)
    """
    # Ensure sources have the same length
    min_len = min([s.shape[0] for s in reference_sources + estimated_sources])
    min_len = min(reference_sources[0].shape[0], estimated_sources[0].shape[0])
    ref_sources_adjusted = np.vstack([s[:min_len] for s in reference_sources])
    est_sources_adjusted = np.vstack([s[:min_len] for s in estimated_sources])

    # Calculate BSS metrics
    sdr, sir, sar, _ = bss_eval_sources(ref_sources_adjusted, est_sources_adjusted)

    return {
        "SDR": sdr.mean(),
        "SIR": sir.mean(),
        "SAR": sar.mean()
    }

## PESQ!!!
def calculate_pesq(reference, processed, fs=8000):
    """Calculate PESQ score between reference and processed audio"""
    try:
        # PESQ expects specific sample rates (8000 or 16000 Hz)
        if fs not in [8000, 16000]:
            raise ValueError("Sample rate must be 8000 or 16000 Hz for PESQ")

        # Convert tensors to numpy arrays
        if isinstance(reference, torch.Tensor):
            reference = reference.cpu().numpy()
        if isinstance(processed, torch.Tensor):
            processed = processed.cpu().numpy()

        # PESQ expects float64 arrays in range [-1, 1]
        reference = reference.astype(np.float64)
        processed = processed.astype(np.float64)

        # Normalize if needed
        if np.max(np.abs(reference)) > 1.0:
            reference = reference / np.max(np.abs(reference))
        if np.max(np.abs(processed)) > 1.0:
            processed = processed / np.max(np.abs(processed))

        # Calculate PESQ
        mode = 'nb' if fs == 8000 else 'wb'
        score = pesq(fs, reference, processed, mode)
        return score
    except Exception as e:
        print(f"Error calculating PESQ: {e}")
        return float('nan')


```

```python
def trim_or_pad_audio(reference_audio, *audio_arrays):
    min_len = len(reference_audio)  #I use the length of the reference as the target
    trimmed_or_padded_audio = []

    for audio in audio_arrays:
        if len(audio) > min_len:
            trimmed_audio = audio[:min_len]  # trim to the minimum length
            trimmed_or_padded_audio.append(trimmed_audio)
        elif len(audio) < min_len:
            padded_audio = np.pad(audio, (0, min_len - len(audio)))  # Pad to the minimum length
            trimmed_or_padded_audio.append(padded_audio)
        else:
            trimmed_or_padded_audio.append(audio)

    return trimmed_or_padded_audio
```

```python
def extract_embedding(audio_path):
    waveform, sr = librosa.load(str(audio_path), sr=16000)
    inputs = feature_extractor(waveform, sampling_rate=16000, return_tensors="pt")
    inputs = {key: val.to(device) for key, val in inputs.items()}
    with torch.no_grad():
        embeddings = base_model(**inputs).embeddings
    return embeddings.cpu().numpy().flatten()
```

### Cosine Similarity Between Original Audios and Separated (from Mixed) Audios

```python

def identify_speakers(original_sources, separated_sources, speaker_ids):
    original_embeddings = [] ##from original
    for source_path in original_sources:
        embedding = extract_embedding(source_path)
        original_embeddings.append(embedding)

    #from separated sources
    separated_embeddings = []
    for source_path in separated_sources:
        embedding = extract_embedding(source_path)
        separated_embeddings.append(embedding)

    # Compute cosine similarity between original and separated embeddings
    similarity_matrix = cosine_similarity(separated_embeddings, original_embeddings)

    # Determine the best match for each separated source
    best_matches = {}
    for sep_idx, sim_scores in enumerate(similarity_matrix):
        # Get the index of the original source with highest similarity
        best_match_idx = np.argmax(sim_scores)
        best_matches[sep_idx] = {
            'original_idx': best_match_idx,
            'speaker_id': speaker_ids[best_match_idx],
            'similarity': sim_scores[best_match_idx]
        }

    return best_matches, similarity_matrix
```

```python
import os
import numpy as np
import torch
import torchaudio
import pandas as pd
from tqdm import tqdm
from pathlib import Path
from transformers import UniSpeechSatModel, Wav2Vec2FeatureExtractor
from sklearn.metrics.pairwise import cosine_similarity
import zipfile
import shutil
import librosa


test_metadata = pd.read_csv("test_metadata.csv")
identification_results = []
correct_identifications = 0
total_samples = 0

for idx, row in tqdm(test_metadata.iterrows(), total=len(test_metadata), desc="Evaluating speaker identification"):

    mixed_path = row['mixed_path']
    original_source1_path = row['source1_path']
    original_source2_path = row['source2_path']
    speaker1_id = row['speaker1_id']
    speaker2_id = row['speaker2_id']

    # Extract the base filename from the mixed path!!
    test_file = os.path.basename(mixed_path)

    # Paths to separated audio files (from previous separation step) in separate_audio folder. names have the mixed path
    source1_path = os.path.join("separated_audio", f"sep1_{test_file}")
    source2_path = os.path.join("separated_audio", f"sep2_{test_file}")


    # Identify speakers
    original_sources= [original_source1_path, original_source2_path]
    separated_sources= [source1_path, source2_path]
    speaker_ids= [speaker1_id, speaker2_id]

    matches, similarity_matrix= identify_speakers(original_sources, separated_sources, speaker_ids)

    # Check if identification was correct
    # For a 2-speaker scenario, we can define correct as:
    # Either (sep1->spk1 and sep2->spk2) OR (sep1->spk2 and sep2->spk1) because we dont know the correspondences
    assignment_1= (matches[0]['original_idx'] == 0 and matches[1]['original_idx'] == 1)
    assignment_2= (matches[0]['original_idx'] == 1 and matches[1]['original_idx'] == 0)

    is_correct = assignment_1 or assignment_2

    result = {
        'file': test_file,
        'speaker1_id': speaker1_id,
        'speaker2_id': speaker2_id,
        'sep1_identified_as': matches[0]['speaker_id'],
        'sep2_identified_as': matches[1]['speaker_id'],
        'sep1_similarity_spk1': similarity_matrix[0][0],
        'sep1_similarity_spk2': similarity_matrix[0][1],
        'sep2_similarity_spk1': similarity_matrix[1][0],
        'sep2_similarity_spk2': similarity_matrix[1][1],
        'correct_identification': is_correct}

    identification_results.append(result)

    if is_correct:
        correct_identifications += 1
    total_samples += 1



# ## Rank-1 accuracy
if total_samples > 0:
    rank1_accuracy = (correct_identifications / total_samples) * 100
    print(f"\nRank-1 Identification Accuracy: {rank1_accuracy:.2f}%")
    print(f"Correctly identified: {correct_identifications} out of {total_samples}")
    results_df = pd.DataFrame(identification_results)
    results_df.to_csv("speaker_identification_results.csv", index=False)
    print("Detailed identification results saved to speaker_identification_results.csv")

    # Additional analysis: per-speaker identification accuracy
    print("\nPer-Speaker Identification Accuracy:")
    for speaker in set(test_metadata['speaker1_id'].tolist() + test_metadata['speaker2_id'].tolist()):
        # Filter results where this speaker is present
        speaker_results = results_df[(results_df['speaker1_id'] == speaker) | (results_df['speaker2_id'] == speaker)]
        if len(speaker_results) > 0:
            speaker_accuracy = (speaker_results['correct_identification'].sum() / len(speaker_results)) * 100
            print(f"Speaker {speaker}: {speaker_accuracy:.2f}% ({speaker_results['correct_identification'].sum()} of {len(speaker_results)})")
else:
    print("No samples were processed successfully.")
```

- Sepformer should be optimizing the separation quality first.
- Cosine Similarity Loss/ Scale-Invariant Signal-to-Distortion Ratio (SI-SDR) Loss can be used for finetuning Sepformer for better speech separation without loss of data or quality.
- Unispeech can be finetuned with LoRA layers and ArcFace loss as already done before.
- In this code, Sepformer wasn't finetuned as there were multiple issues in its training. However, the pipeline has been discussed extensively in the report.
- I use the finetuned Unispeech-sat and pretrained Sepformer models for this task.
- The Sepformer is used for speaker separation and enhancement. On finetuning with the VoxCeleb2 dataset, it would provide even better representations.
- The Unispeech-sat is used for speaker identification between the separated audios of Sepformer and the original audios. The performances of both the simple pre-trained Unispeech-sat model and the finetuned one are reported.

```python
import os
import torch
import torchaudio
import numpy as np
import pandas as pd
from tqdm import tqdm
import soundfile as sf
import librosa
import zipfile
from pesq import pesq
from mir_eval.separation import bss_eval_sources
from speechbrain.inference.separation import SepformerSeparation
from transformers import Wav2Vec2FeatureExtractor, UniSpeechSatForXVector
from peft import LoraConfig, get_peft_model
from sklearn.metrics.pairwise import cosine_similarity


#extract embeddings from pretrained unispeech
def extract_embedding(audio_path, model):
    waveform, sr = librosa.load(audio_path, sr=16000)
    inputs = feature_extractor(waveform, sampling_rate=16000, return_tensors="pt").to(device)
    with torch.no_grad():
        return model(**inputs).embeddings.cpu().numpy().flatten()

```

```python
def evaluate_pair(mixed_path, source1_path, source2_path):
    # Separate audio with SepFormer
    mixed, sr = torchaudio.load(mixed_path)
    if sr != 8000:
        mixed = torchaudio.transforms.Resample(sr, 8000)(mixed)
    temp_path = "temp_mixed.wav"
    torchaudio.save(temp_path, mixed, 8000)

    with torch.no_grad():
        enhanced = sepformer.separate_file(temp_path)
    os.remove(temp_path)

    # Save enhanced audio
    base_name = os.path.basename(mixed_path)
    enh1_path = f"enhanced_audio/enh1_{base_name}"
    enh2_path = f"enhanced_audio/enh2_{base_name}"
    torchaudio.save(enh1_path, enhanced[:, :, 0].cpu(), 8000)
    torchaudio.save(enh2_path, enhanced[:, :, 1].cpu(), 8000)

    #Calculate separation metrics
    orig1, _ = librosa.load(source1_path, sr=16000)
    orig2, _ = librosa.load(source2_path, sr=16000)
    enh1, _ = librosa.load(enh1_path, sr=16000)
    enh2, _ = librosa.load(enh2_path, sr=16000)

    min_len = min(len(orig1), len(orig2), len(enh1), len(enh2))
    ref = np.vstack([orig1[:min_len], orig2[:min_len]])
    est = np.vstack([enh1[:min_len], enh2[:min_len]])

    sdr, sir, sar, _ = bss_eval_sources(ref, est)
    pesq1 = pesq(16000, orig1[:min_len], enh1[:min_len], 'wb')
    pesq2 = pesq(16000, orig2[:min_len], enh2[:min_len], 'wb')

    # Speaker identification (both models)
    # Get embeddings
    orig1_emb = extract_embedding(source1_path, pretrained_model)
    orig2_emb = extract_embedding(source2_path, pretrained_model)

    # Pre-trained model
    pretrained_enh1 = extract_embedding(enh1_path, pretrained_model)
    pretrained_enh2 = extract_embedding(enh2_path, pretrained_model)
    pretrained_sim = cosine_similarity(
        [pretrained_enh1, pretrained_enh2],
        [orig1_emb, orig2_emb]
    )

    # Fine-tuned model
    finetuned_enh1 = extract_embedding(enh1_path, finetuned_model)
    finetuned_enh2 = extract_embedding(enh2_path, finetuned_model)
    finetuned_sim = cosine_similarity(
        [finetuned_enh1, finetuned_enh2],
        [orig1_emb, orig2_emb]
    )

    # Determine correct matches (anyone of the pair could be the match)
    pretrained_correct = (np.argmax(pretrained_sim[0]) == 0 and np.argmax(pretrained_sim[1]) == 1) or \
                         (np.argmax(pretrained_sim[0]) == 1 and np.argmax(pretrained_sim[1]) == 0)

    finetuned_correct = (np.argmax(finetuned_sim[0]) == 0 and np.argmax(finetuned_sim[1]) == 1) or \
                        (np.argmax(finetuned_sim[0]) == 1 and np.argmax(finetuned_sim[1]) == 0)

    return {
        'file': base_name,
        'SDR': np.mean(sdr),
        'SIR': np.mean(sir),
        'SAR': np.mean(sar),
        'PESQ': (pesq1 + pesq2)/2,
        'pretrained_correct': pretrained_correct,
        'finetuned_correct': finetuned_correct,
        'pretrained_sim_matrix': pretrained_sim.tolist(),
        'finetuned_sim_matrix': finetuned_sim.tolist()
    }

```

```python
test_df = pd.read_csv("test_metadata.csv")
os.makedirs("enhanced_audio", exist_ok=True)

results = []
pretrained_correct = 0
finetuned_correct = 0

for _, row in tqdm(test_df.iterrows(), total=len(test_df)):
    result = evaluate_pair(
        row['mixed_path'],
        row['source1_path'],
        row['source2_path']
    )
    results.append(result)
    pretrained_correct += int(result['pretrained_correct'])
    finetuned_correct += int(result['finetuned_correct'])
```

```python
results_df = pd.DataFrame(results)
results_df.to_csv("evaluation_results.csv", index=False)
avg_sdr = results_df['SDR'].mean()
avg_sir = results_df['SIR'].mean()
avg_sar = results_df['SAR'].mean()
avg_pesq = results_df['PESQ'].mean()

pretrained_acc = pretrained_correct / len(test_df) * 100
finetuned_acc = finetuned_correct / len(test_df) * 100
```

```python
print("\n=== Final Results ===")
print("Separation Performance:")
print(f"SDR: {avg_sdr:.2f} dB | SIR: {avg_sir:.2f} dB | SAR: {avg_sar:.2f} dB")
print(f"PESQ: {avg_pesq:.2f}")
```
